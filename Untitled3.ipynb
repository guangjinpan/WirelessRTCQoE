```python
data_train=np.zeros((1,20))
data_test=np.zeros((1,20))
for i in range(1,8):
    j=1
    file_name1 =f'../data/UE2_{i}_{j}_train.csv'
    datax=pd.read_csv(file_name1,index_col=0).values  
    data_train=np.vstack((data_train,datax))
    file_name1 =f'../data/UE2_{i}_{2}_test.csv'
    datax=pd.read_csv(file_name1,index_col=0).values  
    data_test=np.vstack((data_test,datax))    
    plt.plot(datax[:,0],datax[:,4]*100)
    plt.plot(datax[:,0],datax[:,11])
#     print(datax[:,11])
    plt.show()
data_train=data_train[1:,:]
data_test=data_test[1:,:]

d=data_train
data_train=data_test
data_test=d
```

```python
data_train_time=np.zeros((len(data_train),35))
data_test_time=np.zeros((len(data_test),35))
data_train0=data_train[:,[2,4,6,7,8,9]]
data_test0=data_test[:,[2,4,6,7,8,9]]
data_train_time[:,:6]=data_train0
data_train_time[1:,6:12]=data_train0[:-1,:]
data_train_time[2:,12:18]=data_train0[:-2,:]
data_train_time[3:,18:24]=data_train0[:-3,:]
data_train_time[4:,24:30]=data_train0[:-4,:]
data_train_time[4:,30]=data_train[3:-1,11]
data_train_time[4:,31]=data_train[2:-2,11]
data_train_time[4:,32]=data_train[1:-3,11]
data_train_time[4:,33]=data_train[0:-4,11]
data_train_time[:,34]=data_train[:,11]


data_test_time[:,:6]=data_test0
data_test_time[1:,6:12]=data_test0[:-1,:]
data_test_time[2:,12:18]=data_test0[:-2,:]
data_test_time[3:,18:24]=data_test0[:-3,:]
data_test_time[4:,24:30]=data_test0[:-4,:]
data_test_time[4:,30]=data_test[3:-1,11]
data_test_time[4:,31]=data_test[2:-2,11]
data_test_time[4:,32]=data_test[1:-3,11]
data_test_time[4:,33]=data_test[0:-4,11]
data_test_time[:,34]=data_test[:,11]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error


# 创建SVR回归模型对象
clf = SVR(kernel='linear')

# 拟合（训练）模型
clf.fit(data_train_time[:,:33], data_train_time[:,34])

# 预测测试集
y_pred_SVR = clf.predict(data_test_time[:,:33])

# 计算MAE
mae = mean_absolute_error(data_test_time[:,34], y_pred_SVR)

print("预测结果：", y_pred_SVR)
print("MAE：", mae)
```

```python
from sklearn.ensemble import RandomForestRegressor

# 创建SVR回归模型对象
clf = RandomForestRegressor()

# 拟合（训练）模型

clf.fit(data_train_time[:,:33], data_train_time[:,34])

# 预测测试集
y_pred_RandomForest = clf.predict(data_test_time[:,:33])

# 计算MAE
mae = mean_absolute_error(data_test_time[:,34], y_pred_RandomForest)

print("预测结果：", y_pred_RandomForest)
print("MAE：", mae)
print("MAE：", np.mean(np.abs(data_test_time[:,34]- y_pred_RandomForest)))
```
